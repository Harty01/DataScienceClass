```python
def missing(col_name):
    if df.dtypes[col_name]=='object':
        replace_val=df[col_name].mode()[0]
    elif df[col_name].std() > df[col_name].mean():    #GENERAL CODE COMBINED
        replace_val=df[col_name].median()
    else:
        replace_val=df[col_name].mean()
    df[col_name].fillna(replace_val, inplace= True)
missing('glucose')
missing('cigsPerDay')
missing('education')
missing('BPMeds')
missing('totChol')
missing('heartRate')
missing('BMI')
```
